```python
# using RegEX
RegEX are used to detect patterns, whenever we want to repalce pattern with some other values then we use regex
```

```python
df = pd.read_csv('D:/Data_analyst/Python/Pandas/weather_data.csv')
df
```

```python
new_df = df.replace('[A-Za-z]','',regex=True)
new_df
#in the below df it replaced all the alphabets with blank and hence the event col is also replaced
```

#### Replacing the list of values with other list of values

```python
df = pd.DataFrame({
    'score': ['Good','Exceptional','Average','Poor','Average'],
    'student':['rob','juliet','jhon','meer','Anu']
})
df
```

```python
new_df = df.replace(['Poor','Average','Good','Exceptional'],[1,2,3,4])
new_df
```
