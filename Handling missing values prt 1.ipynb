```python
#parse_dates will convert the given column into date
df = pd.read_csv('D:/Data_analyst/Python/Pandas/weather_data.csv',parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
# to set date as index
df.set_index('day',inplace=True)
```

```python
df
```

```python
#fillna method
#the below method replaced all the NaN values with 0
new_df = df.fillna(0)
new_df
```

```python
#to fill specfic value for specific col with fillna method
new_df = df.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'event'  : 'No Event'
})
new_df
```

```python
# to fill the NA values by using the forward fill (ffill) method
#carry forward method is nothing but using the previous days values
new_df = df.fillna(method='ffill')
new_df
```

```python
# to fill the NA values by using the backward fill (bfill) method
#backward fill method will fill the missing values with  next days values
new_df = df.fillna(method='bfill')
new_df
```

```python
#fillna method has the follwing arguments
axis=0 bydefault means replace values vertically 
axis=1 replace values horizontally ()
```

```python
new_df = df.fillna(method='bfill',axis=1)
new_df
```

```python
#fillna method has the follwing arguments
limit=1 then it will only copy the value once 
#see the belew example on date 01-07 and 01-08
```

```python
new_df = df.fillna(method='ffill',limit=1)
new_df
```

##### Interpolate method in Pandas

```python
Interpolate method works on liner interpolation to guess the missing values between 2 cells
before inerpoation
32 NA 28

after interpolation
32 30 28
```

```python
df
```

```python
new_df = df.interpolate()
new_df
```

```python
#using the time argument in interpolate function
#if you look cosely in the original df for 1st 3 cells then you will notice 04th is not in between 01 and 05 in day column
#hence its value should be more close to 05
#using the time argument we can make this possible as it will consider time also
```

```python
new_df = df.interpolate(method='time')
new_df
```

##### Drop NA method

```python
#it will drop all the rows which has aleast one NA value
new_df = df.dropna()
new_df
```

```python
#how method in dorp na
#it will drop only those rows which has all the cells values as NA
new_df  = df.dropna(how = 'all')
new_df
```

```python
#thresh method
#it will drop the rows based on no.of valid values
#thersh=1 means it will drop the rows in which cells has only has one valid data
#thersh=2 means it will drop the rows in which cells has only has two valid data
```

```python
new_df  = df.dropna(thresh = 1)
new_df
```

```python
new_df  = df.dropna(thresh = 2)
new_df
```
