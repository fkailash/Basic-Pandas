## Append 2 data frames

```python
##append means  - stacking tow df on top of each other
Append to add new rows
eg. adding monthly data
```

```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    'city' : ['mumbai','delhi','banglore'],
     'temperature' : [32,45,30],
    'humidity' : [80,60,78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    'city' : ['new york','chicago','orlando'],
     'temperature' : [21,14,35],
    'humidity' : [68,65,75]
})
us_weather
```

```python
#to combine two dataframes
#created a dataframe from the above 2 dataframes
df = pd.concat([india_weather,us_weather])
df
```

```python
#to provide a continuous index
df = pd.concat([india_weather,us_weather], ignore_index=True)
df
```

```python
#the key argument will assign a key for each of the dataframe that we have passed in the conact func
#keys also help us to retrive original data sets when you merged you ds into a larger data set
df = pd.concat([india_weather,us_weather],keys=['india','us'])
df
```

```python
#now to retrive a single dataset then 
#loc stands for loaction
df.loc['india']
```

```python
df.loc['us']
```

# Merge

```python
df1 = pd.DataFrame({
    'city' : ['new york','chicago','orlando'],
     'temperature' : [21,14,35]
})
df1
```

```python
df2 = pd.DataFrame({
    'city' : ['new york','chicago','orlando'],
     'humidity' : [68,65,75]
})
df2
```

```python
# to merge the above df we can use merge option
df3 = pd.merge(df1,df2,on = 'city')
df3
```

```python
## Suppose you don't have consistent data
```

```python
df1 = pd.DataFrame({
    'city' : ['new york','chicago','orlando','baltimore'],
     'temperature' : [21,14,35,32]
})
df1
```

```python
df2 = pd.DataFrame({
    'city' : ['chicago','new york','san franciso'],
     'humidity' : [65,68,71]
})
df2
```

```python
# the below code performed an inner join and returned only the matching elements
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
#the how agrument in merge will help us to define the type of join
df3 = pd.merge(df1,df2,on='city',how = 'inner')
df3
```

```python
#the how agrument in merge will help us to define the type of join
df3 = pd.merge(df1,df2,on='city',how = 'outer')
df3
```

```python
#left join - it will return the left df data and the matching elements in both
df3 = pd.merge(df1,df2,on='city',how = 'left')
df3
```

```python
#right join - it will return the right df data and the matching elements in both
df3 = pd.merge(df1,df2,on='city',how = 'right')
df3
```

```python
#the indicator argument will allow you to know from which df the data is coming
df3 = pd.merge(df1,df2,on='city',how = 'outer',indicator = True)
df3
```

```python
df1 = pd.DataFrame({
    'city' : ['new york','chicago','oroland','baltimore'],
     'temperature' : [32,45,30,38,],
    'humidity' : [80,60,78,80]
})
df1
```

```python
df2 = pd.DataFrame({
    'city' : ['chicago','new york','san deigo'],
     'temperature' : [21,14,35],
    'humidity' : [68,65,75]
})
df2
```

```python
#since it has same column name pandas will automatically appned suffixes as x and y
df3 = pd.merge(df1,df2, on='city')
df3
```

```python
#providing our own sffixes
df3 = pd.merge(df1,df2, on='city',suffixes = ('_left','_right'))
df3
```
